```python
df = pd.read_excel(
    'Results.xlsx',
    sheet_name='Table 1',
    header=1
)
```

```python
df.head()
df.columns
```

```python
df = df.dropna(axis=1, how='all')
```

```python
df = df.rename(columns={
    'CANDIDATE NAME': 'candidate',
    'GENDER': 'gender',
    'AGE': 'age',
    'CATEGORY': 'category',
    'PARTY': 'party',
    'TOTAL VALID VOTES POLLED': 'votes'
})
```

```python
# Convert age and votes columns to numeric (handle cases where columns might have different names)
if 'age' in df.columns:
    df['age'] = pd.to_numeric(df['age'], errors='coerce')
if 'votes' in df.columns:
    df['votes'] = pd.to_numeric(df['votes'], errors='coerce')
```
